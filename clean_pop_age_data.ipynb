```python
import pandas as pd
import numpy as np
```

I spent some time searching for a single dataset that included historical data for the age composition of the US. I'm sure there's one out there, but in the end, the best I could do were three separate sets of data from 1990-2000, 2000-2010 and 2010-2015 (I didn't go back further because my employment by education level data starts in 1992). Note that the data sets included age composition by gender as well, but I discard it below because it was beyond the scope of what my questions.

```python
#Helper function to parse the 1990-2000 data
def readPopAgeFile(path, startLine, years):
    return pd.read_table(path, sep='\s{2,}', nrows=21, skiprows=startLine, names=years)

#in the file for the first half of the decade was separated from the data from data for the last five,
#so we collect both parts then concatenate them
start95 = readPopAgeFile('1990_age_pop.txt', 12, ['2000', '1999', '1998', '1997', '1996', '1995'])
start90 = readPopAgeFile('1990_age_pop.txt', 190, ['1994', '1993', '1992', '1991', 'todrop', '1990'])
pop_age_1990 = pd.concat([start90, start95], axis=1).sort_index(axis=1).drop('todrop', axis=1)

pop_age_1990.head()
```

```python
# read in the data for 2000-2010
pop_age_2000 = pd.read_csv('2000_age_pop.csv')
pop_age_2000.head()
```

```python
#Various transformations to make get the data in a form we want
pop_age_2000.drop(['April 1, 20001', 'April 1, 20102'], axis=1, inplace=True)
pop_age_2000.loc[0, 'July 1, 20103'] = 2010
pop_age_2000 = pop_age_2000.set_index('Sex and Age')
pop_age_2000.index.name=None
pop_age_2000.columns = pop_age_2000.iloc[0, :]
pop_age_2000 = pop_age_2000[pop_age_2000['2000'] != '2000']
pop_age_2000.columns.name = None
pop_age_2000.head()
```

```python
#read in the 2010 data
pop_age_2010 = pd.read_excel('2010_pop_age.xls', skiprows=4, skip_footer=24)
pop_age_2010.head()
```

```python
pop_age_2010['age_group'] = pop_age_2010['Unnamed: 0']
toDrop  = [c for c in pop_age_2010.columns if "Male" in c or "Female" in c or 'Unnamed' in c]
pop_age_2010.drop(toDrop, inplace=True, axis=1)
pop_age_2010.drop(['Both Sexes', 'Both Sexes.1'], axis=1, inplace=True)
pop_age_2010.columns = ['2010', '2011', '2012', '2013', '2014', '2015', 'age_group']
pop_age_2010.set_index('age_group', inplace=True)
pop_age_2010.index.name= None
pop_age_2010.head()
```

```python
#The datasets overlap by a year (ie. both the 1990s data and 2000s data contain the year 2000)
#So we drop the overlapping year from the earlier datasets.
pop_age_1990.drop('2000', axis=1, inplace=True)
pop_age_2000.drop(2010, axis=1, inplace=True)
```

```python
#The convert the data from strings to ints. Additionally, we rescale the 1990's
# data, as it originally was in 1000s while the data from the other two decades was not. 
def convertToNum(x):
    x =x.replace(",", "")
    return int(x)

pop_age_1990 = pop_age_1990.applymap(convertToNum)
pop_age_1990 = pop_age_1990.applymap(lambda x: x * 1000)

pop_age_2000 = pop_age_2000.applymap(convertToNum)

pop_age_2010 = pop_age_2010.applymap(convertToNum)
pop_age_2010.head()
```

```python
#The oldest age range for the 2010s data was 85 and older,
#while the other two datasets went up to 100. So we create
#an 85 and older category for the other two and drop the older
#categories
def create85AndOlder(df):
    older85  = df.iloc[-4:, :].sum()
    older85.name = '85 years and over'
    df.drop(df.index.values[-4:], inplace=True)
    return df.append(older85)

pop_age_2000 = create85AndOlder(pop_age_2000)
pop_age_1990 = create85AndOlder(pop_age_1990)
```

```python
pop_age_2000
```

```python
#We standardize the indices of the datasets so that we can concatenate them
def standardizeIndex(ind):
    index = [i.lower().strip().replace(" ", "_").replace(".", "") for i in ind.values]
    return index
pop_age_1990.index = standardizeIndex(pop_age_1990.index)
pop_age_2000.index = standardizeIndex(pop_age_2000.index)
pop_age_2010.index = standardizeIndex(pop_age_2010.index)
```

```python
all_years = pd.concat([pop_age_1990, pop_age_2000, pop_age_2010], axis=1).dropna()
```

```python
all_years.to_csv('all_years_age_pop.csv', index=True, index_label='age_group', header=True)
```
